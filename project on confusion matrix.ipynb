```python
import numpy as np #used for matrix
from sklearn.metrics import confusion_matrix 
import seaborn as sns
import matplotlib.pyplot as plt #used to plot a matrix
```

```python
actual = np.array(['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted= np.array(['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])

cm = confusion_matrix(actual,predicted)

sns.heatmap(cm, annot = True, fmt='g', xticklabels=['Dog','Not Dog'], yticklabels=['Dog','Not Dog']) #fmt is a flat matrix used to conver in categorical or in string value, heatmap or heatmatrix is a prediction or it is used to show results in positive and negative, annot is uesd to display the numeric or string values in heatmapcells
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```
