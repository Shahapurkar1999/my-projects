```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris #load_iris consist of leaf data and it has some
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.feature_names # contents of leaf data
```

```python
iris.target_names # 
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names) #checking feature names
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
#for sepal length
plt.figure(figsize=(10,8))
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
```

```python
#for petal length
plt.figure(figsize=(10,8))
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'],axis='columns')
y = df.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2)
```
